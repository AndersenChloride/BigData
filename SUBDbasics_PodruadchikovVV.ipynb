##Вопросы

1. С помощью какой команды можно посмотреть все таблицы в базе данных?  
Ответ: 2) \dt
2. При помощи какой команды можно получить краткую справку по всем сервисным командам?  
Ответ: 4) \\?
3. Как отобразить таблицу airports а базе данных "Пассажирские авиаперевозки"?  
Ответ: SELECT * FROM airports;
4. С помощью каких ключевых слов можно создать таблицу?  
Ответ: CREATE TABLE.
5. С помощью каких ключевых слов можно сгруппировать таблицу?  
Ответ: GROUP BY.
6. С помощью каких ключевых слов можно отсортировать таблицу?  
Ответ: ORDER BY.
7. С помощью каких ключевых слов можно обновить таблицу?  
Ответ: UPDATE.
8. С помощью какого ключевого слова можно посчитать количество какого-то атрибута?  
Ответ: COUNT.
9. С помощью какой команды можно подключиться к базе данных demo с именем пользователя postgres?  
Ответ: 1) psql -d demo -U postgres
